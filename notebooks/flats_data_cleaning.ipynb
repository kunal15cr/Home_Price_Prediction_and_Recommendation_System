```python
def clean_data(df):
    return (
        df.drop(['property_id','link','property_id'],axis=1)
        .loc[df['price'] != 'Price on Request']
        .assign(
            society=lambda df_: df_.society.apply(
                lambda x: re.sub(r'\d+\.\d+ ★', '', str(x)).strip().lower()
                                                  ),
            #price = lambda df_: df_.price.pipe()
        )

    )
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```
