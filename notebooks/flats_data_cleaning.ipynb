```python
def treat_price(x):
    # If x is already a numeric type, return it as is (rounded)
    if isinstance(x, (float, int)):
        return round(x, 2)
    
    # Handle non-numeric or malformed strings gracefully
    try:
        # Split the string by space (e.g., "100 Lac")
        x = x.split(" ")

        # If there's a unit, check for 'Lac' and process accordingly
        if len(x) == 2:
            if x[1] == 'Lac':
                return round(float(x[0]) / 100, 2)
            else:
                return round(float(x[0]), 2)
        else:
            # If no unit, simply return the float value of the first part
            return round(float(x[0]), 2)
    except ValueError:
        # If the conversion fails (e.g., if 'Price' is encountered), return None or some default value
        print(f"Skipping invalid input: {x}")
        return None  # Or you could return a default value, like 0 or an empty string
```

```python
import re

def clean_data(df):
    return (
        df.drop(['property_id', 'link'], axis=1)
        .loc[(df['price'] != 'Price on Request') & (~df['bedRoom'].isnull())]
        .assign(
            society=lambda df_: df_.society.apply(
                lambda x: re.sub(r'\d+\.\d+ ★', '', str(x)).strip().lower()  # Clean 'society'
            ),
            price=lambda df_: df_.price.apply(treat_price),
            area = lambda df_: df_.area.str.split("/").str.get(0).str.replace(",","").str.replace("₹",""),
            bedRoom = lambda df_: df_.bedRoom.str.split(" ").str.get(0).astype(int),
            bathroom = lambda df_: df_.bathroom.str.split(" ").str.get(0).astype(int),
            balcony = lambda df_: df_.balcony.str.split(' ').str.get(0).str.replace('No','0'),
            additionalRoom = lambda df_: df_.additionalRoom.fillna('not available').str.lower(),
            floorNum = lambda df_:df_.floorNum.str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
        )
    ).rename(columns={'area':'price_per_sqft'})

# Example call to clean the data
clean_data(df=flats_data).head()
```
